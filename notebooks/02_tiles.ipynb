```python
from geomltoolkits.downloader import tms as TMSDownloader
import geopandas as gpd
```

```python
data = gpd.read_file("../data/trees.geojson")
```

```python
data.total_bounds
```

```python
data.to_crs(epsg=4326,inplace=True)
data.total_bounds
```

```python
min_lon, min_lat, max_lon, max_lat = data.total_bounds
```

```python
bbox = [min_lon, min_lat, max_lon, max_lat]
```

```python
import os 
ZOOM = 19
WORK_DIR = "../data"
TMS = "https://tiles.openaerialmap.org/5a28639331eff4000c380690/0/5b1b6fb2-5024-4681-a175-9b667174f48c/{z}/{x}/{y}.png"

os.makedirs(WORK_DIR, exist_ok=True)

await TMSDownloader.download_tiles(
    tms=TMS,
    zoom=ZOOM,
    out=WORK_DIR,
    bbox=bbox,
    georeference=True,
    dump_tile_geometries_as_geojson=True,
    prefix="OAM"
)
```

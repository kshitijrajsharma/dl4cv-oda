```python
import geopandas as gpd 
```

```python
data = gpd.read_file("../data/raw/kolovai-trees.geojson")
```

```python
data
```

```python
def string_to_dict(x):
    dict_ = {}
    pairs = x.split(',')
    for pair in pairs:
        if pair: 
            key, value = pair.split('=>')
            dict_[str(key.strip('"'))] = str(value.strip('"'))
    return dict_
```

```python
data.dropna(subset=['other_tags'], inplace=True)
```

```python
data['tags_dict'] = data['other_tags'].apply(string_to_dict)
```

```python
data 
```

```python
data['natural'] = data['tags_dict'].apply(lambda x: x.get('natural', None))
data['species'] = data['tags_dict'].apply(lambda x: x.get('species', None))
```

```python
data
```

```python
data['natural'].value_counts()
```

```python
data['species'].value_counts()
```

```python
mapping = {'Mangifera indica':'Mango',
           'Mangifera indica L.':'Mango',
           'Cocos nucifera':'Coconut',
           'Cocus nucifera':'Coconut',
           'cocos nucifera':'Coconut',
           'Cocos nucifera L.': 'Coconut',
           'Cocos c':'Coconut',
           'Carica tree':'Papaya',
           'Carica papaya':'Papaya',
           'Musaceae':'Banana',
           'Musa':'Banana',}
```

```python
len(mapping.keys())
```

```python
data['species_mapped'] = data['species'].map(mapping)
```

```python
data['species_mapped'].value_counts()
```

```python
data.columns
```

```python
data = data[['osm_id', 'natural', 'species', 'species_mapped','geometry']]
```

```python
data
```

```python
data.dropna(subset=['species_mapped'], inplace=True)
```

```python
data
```

```python
data['species_mapped'].value_counts()
```

```python
data.to_crs(epsg=3857, inplace=True)
```

```python
from shapely.geometry import box

buffered_gdf = data.copy()

buffered_gdf["lon"] = buffered_gdf.geometry.x
buffered_gdf["lat"] = buffered_gdf.geometry.y


buffered_gdf['geometry'] = buffered_gdf['geometry'].buffer(3) # 3 meters buffer 


buffered_gdf["geometry"] = buffered_gdf.geometry.apply(
    lambda g: box(*g.bounds)
)
```

```python
data.to_file("../data/raw/kolovai-trees-cleaned.geojson", driver='GeoJSON')
```
